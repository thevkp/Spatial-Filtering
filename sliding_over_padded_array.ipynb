````python
import numpy as np
import matplotlib.pyplot as plt
import cv2

plt.figure(figsize=(5, 5))
def print_all_window(arr, k):
  window = []
  for i in range(len(arr) - k + 1):
    window.append(np.array(arr[i : i + k])) # This is where you sliding
    plt.subplot(1, len(arr) - k + 2, i + 1)
    plt.plot(window[i])
    plt.title(i)
    plt.axis('off')
  
  plt.show()
  
  return window
    
  
arr = np.random.randint(0, 25, 35)
windows = print_all_window(arr, 3)
print(windows)
````

````python
def averaging(arr, k):
  pad_width = k // 2
  padded = np.pad(arr, ((pad_width, pad_width)), mode='edge')
  
  averaged = arr.copy().astype(float)
  
  avg_window = []
  for i in range(len(arr)):
    window = padded[i : i + k] # This where kernel is sliding 
    avg = np.mean(window)
    avg_window.append(avg)
    averaged[i] = avg # Average of the neighboring pixels applied to center pixel

  return (averaged, avg_window)

arr = np.random.randint(0, 256, 360)
# print(arr)
averaged_array, avg_window = averaging(arr, 15)
# print(averaged_array)
# print(avg_window)

````
