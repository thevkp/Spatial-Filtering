```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

arr = np.random.randint(0, 256, 225)
arr = arr.reshape(15, 15)
# print(arr)
```

```python
def pad_arr(img, pad):
  return np.pad(img, ((pad, pad), (pad, pad)), mode='edge')

pad_arr = pad_arr(arr, 1).astype(np.float32)
# print(pad_arr)
```

```python
kernel = np.ones((3, 3), np.float32) / 9.0

filetered_arr = cv2.filter2D(pad_arr, -1, kernel)

# print(filetered_arr)
```

```python
# arr = np.sort(arr)
# pad_arr = np.sort(pad_arr)
plt.figure(figsize=(8, 6), dpi=150)

plt.subplot(1, 2, 1)
plt.imshow(arr, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.imshow(pad_arr, cmap='gray', vmin=0, vmax=255)
plt.show()
```
